```python
import os
import glob
import json

# 카테고리 답변 폴더 경로
answer_path = './healthcare_data/Training/TL/2.답변/응급질환/두통'

# 모든 JSON 파일 경로 가져오기
answer_files = glob.glob(os.path.join(answer_path, '**', '*.json'), recursive=True)

# JSON 파일 읽기
answer_data = []
for file_path in answer_files:
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
        answer_data.append(data)
```

```python
# 질문 데이터 열기
question_path = './healthcare_data/Training/TL/1.질문/응급질환/두통'
question_files = glob.glob(os.path.join(question_path, '**', '*.json'), recursive=True)

# JSON 파일 읽기
question_data = []
for file_path in question_files:
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
        question_data.append(data)
```

### 답변 데이터프레임으로 만들기

```python
import pandas as pd

df = pd.DataFrame(answer_data)

df.head(2)
```

```python
# disease_name 컬럼안에 있는 데이터가 dict 형태여서 바로 키 값을 불러오는 코드 작성
df['disease_name_kor'] = df['disease_name'].apply(lambda x: x['kor'])
```

```python
# 답변 한줄로 병합
df['answer_intro_conclusion'] = df['answer'].apply(lambda x: f"{x['intro']} {x['conclusion']}")
df['answer_all'] = df['answer'].apply(lambda x: f"{x['intro']} {x['body']} {x['conclusion']}")

df['answer_intro_conclusion'][5513]
```

```python
df.head(3)
```

```python
answer_df = df[['disease_category', 'disease_name_kor', 'intention', 'answer_intro_conclusion', 'answer_all']]
```

```python
answer_df.rename(columns = {'disease_name_kor': 'disease_name', 'answer_all': 'answer'}, inplace=True)
```

```python
answer_df
```

```python
head_answer_df = answer_df[(answer_df['intention'] == '증상') | (answer_df['intention'] == '진단')].reset_index(drop=True)
head_answer_df
```

```python
import numpy as np

# 증상, 진단 랜덤으로 50개씩 추출
head_filtered_answer1_df = head_answer_df[head_answer_df['intention'] == '증상']
head_filtered_answer2_df = head_answer_df[head_answer_df['intention'] == '진단']

head_100_answer_df = pd.concat([head_filtered_answer1_df.sample(n=50), head_filtered_answer2_df.sample(n=50)])
head_100_answer_df = head_100_answer_df.reset_index(drop=True)
head_100_answer_df
```

```python
head_100_answer_df.to_csv('두통 답변 100.csv')
```

### 질문 데이터프레임으로 만들기

```python
df = pd.DataFrame(question_data)

df.head(2)
```

```python
# disease_name 컬럼안에 있는 데이터가 dict 형태여서 바로 키 값을 불러오는 코드 작성
df['disease_name_kor'] = df['disease_name'].apply(lambda x: x['kor'])
```

```python
question_df = df[['disease_category', 'disease_name_kor', 'intention', 'question']]
question_df.rename(columns = {'disease_name_kor': 'disease_name', 'answer_a': 'answer'}, inplace=True)
question_df
```

```python
filtered_question_df = question_df[(question_df['intention'] == '증상') | (question_df['intention'] == '진단')].reset_index(drop=True)
filtered_question_df
```

```python
for i, text in enumerate(filtered_question_df['question']):
    if '진단 절차' in text:
        print(i, text)
```
